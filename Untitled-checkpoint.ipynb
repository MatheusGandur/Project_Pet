```python
df['created_at'] = pd.to_datetime(df['created_at'])
df['deleted_at'] = pd.to_datetime(df['deleted_at'])
```

```python
df['Deletado'] = df['deleted_at'].notnull()
```

```python
entradas_por_mes = df.groupby(df['created_at'].dt.to_period('M')).size()
```

```python
saidas_por_mes = df.groupby(df['deleted_at'].dt.to_period('M')).size()
```

```python
entradas_por_mes = entradas_por_mes.reindex(saidas_por_mes.index, fill_value=0)
```

```python
dados = pd.DataFrame({'Entradas': entradas_por_mes, 'Saidas': saidas_por_mes})
```

```python
ax = dados.plot(kind='bar', stacked=True, figsize=(10, 6), color=['green', 'red'])
```

```python
plt.xlabel('Mês')
plt.ylabel('Número de Registros')
plt.title('Entradas e Saídas por Mês')
```

```python
plt.show()
```

```python
entradas_por_mes = df[df['Deletado'] == False].groupby(df['created_at'].dt.to_period('M')).size()
saidas_por_mes = df[df['Deletado'] == True].groupby(df['deleted_at'].dt.to_period('M')).size()
ativos_por_mes = df[df['status'] == 'active'].groupby(df['created_at'].dt.to_period('M')).size()
cancelados_por_mes = df[df['status'] == 'canceled'].groupby(df['deleted_at'].dt.to_period('M')).size()
```

```python
entradas_por_mes = entradas_por_mes.reindex(saidas_por_mes.index, fill_value=0)
ativos_por_mes = ativos_por_mes.reindex(saidas_por_mes.index, fill_value=0)
cancelados_por_mes = cancelados_por_mes.reindex(saidas_por_mes.index, fill_value=0)
```

```python
dados = pd.DataFrame({'Entradas': entradas_por_mes, 'Saidas': saidas_por_mes, 'Ativos': ativos_por_mes, 'Cancelados': cancelados_por_mes})
```

```python
ax = dados.plot(kind='bar', stacked=True, figsize=(12, 8), color=['green', 'red', 'blue', 'orange'])
```

```python
plt.xlabel('Mês')
plt.ylabel('Número de Registros')
plt.title('Entradas, Saídas, Ativos e Cancelados por Mês')
```

```python
plt.show()
```
